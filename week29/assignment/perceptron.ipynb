```python
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd
from sklearn.model_selection import train_test_split

raw_data = pd.read_csv("../../customer_purchase_data.csv")
print(raw_data.head())

X= raw_data[['Age','EstimatedSalary']]
Y= raw_data['Purchased']
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)

clf = Perceptron(
    max_iter=100,      
    eta0=0.3,        
    random_state=42, 
    tol=1e-3, 
    verbose=1
)

clf.fit(X_train, Y_train)

print("\n── Learned parameters ──")
print(f"  Weight for x1 (w1): {clf.coef_[0][0]:.4f}")
print(f"  Weight for x2 (w2): {clf.coef_[0][1]:.4f}")
print(f"  Epochs run:         {clf.n_iter_}")

Y_pred = clf.predict(X_test)

print("\n── Predictions ──")
for inputs, pred, label in zip(X_test, Y_pred, Y_test):
    status = "✓" if pred == label else "✗"
    print(f"  {status}  x={inputs}  predicted={pred}  label={label}")

print("\n── Accuracy ──")
print(f"  {accuracy_score(Y_test, Y_pred) * 100:.1f}%")
print(classification_report(Y_test, Y_pred))
```
